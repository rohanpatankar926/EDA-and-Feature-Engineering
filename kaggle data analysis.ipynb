### Let's see the Programming and the recomended languages

```python
def col(q,c):
    x = []
    for i in range(1,c+1):
        st = f'Q{q}_Part_{i}'
        x.append(st)
    return x
```

```python
data2=col(7,12)
d=pd.read_csv("kaggle_survey_2021_responses.csv",skiprows=[1],usecols=data2)
```

```python
def merge(x,itr,no,string='None'):
    merged = {}
    for i in range(1,itr):
        merged.update(dict(x[f'Q{no}_Part_{i}'].value_counts()))
    merged1 = pd.DataFrame(merged.items(), columns = [string, 'counts'])
    return merged1
```

```python
language=merge(d,itr=13,no=7,string="Prog language")
```

```python
language
```

### Which language kagglers used on regular basis ?

```python
plt.figure(figsize=(10,5))
x=language["Prog language"]
y=language["counts"]
ax=sns.barplot(x=x,y=y)
plt.title("most programming language used by kagglers",fontsize=15)
ax.set(xlabel="programming languages")
plt.show()
```
