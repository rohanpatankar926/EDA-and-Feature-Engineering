```python
#here i needed country code on left table i.e data named dataset so iam using left join on country code
final_data=pd.merge(data,data1,on="Country Code",how="left")
```

```python
ratings=final_data.groupby([ 'Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"ratings counts"})
```

```python
ratings.max()
```

```python
ratings
```

```python
#observation 
#when ratings is b/n 4.5 to 4.9-->excellent
#when ratings is b/n 3.5 to 3.9-->good
#when ratings is b/n 3.5 to 3.9-->good
#when ratings is b/n 2.5 to 3.4-->avg

#when ratings is b/n 0.0 to 2.4-->poor
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=ratings["Aggregate rating"],y=ratings["ratings counts"])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=ratings["Aggregate rating"],y=ratings["ratings counts"],hue=ratings["Rating color"]
           ,palette=["grey","red","orange","yellow","green","green"])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,4))
cp=sns.countplot(x="Rating color", data=ratings,palette=["grey","red","orange","yellow","green","green"])
for p in cp.patches:
    cp.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
```

```python
##find the country name that has not given 0 ratings

zero_ratings=final_data[final_data["Aggregate rating"]==0].groupby("Country").size().reset_index().rename(columns={0:"0 ratings"})
zero_ratings
```
